```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load the data into a Pandas DataFrame
df = pd.read_excel("./dim_carbon_footprint_country.xlsx")
# Calculate the per capita carbon footprint by dividing the total carbon footprint by the population
df['per_capita_carbon_footprint'] = (df['carbo_footprint'] / df['population']) *100000

# Print the resulting DataFrame
df.head()
```
